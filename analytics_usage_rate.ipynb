# ブキ・サブ・スペシャルの使用率を可視化する

```python
import datetime as dt
import src.analytics2 as a
import src.visualize2 as v
import src.definitions as d

# Context
date_from = dt.date(2022, 12, 19)
date_to = date_from + dt.timedelta(days=7)
lobby = d.Lobby.XMATCH

f"{str(date_from)} <= date < {str(date_to)}"
```

```python
translations = v.get_translations()
details = a.read_details_from_to(date_from, date_to, lobby)
players = a.details_to_players(details)
details["mode"].value_counts()
```

```python
# ブキ使用率

aggregated = a.aggregate_index_per_subject(players, "weapon", "usage-rate")
plt, ax = v.show_aggregated_heatmap(
    aggregated,
    title=f"{translations[lobby.value]} ブキ使用率 [%]",
    figsize=(6, 28),
    use_annotation_image=True,
)
plt.show()
```

```python
# サブ使用率

aggregated = a.aggregate_index_per_subject(players, "weapon-sub", "usage-rate")
plt, ax = v.show_aggregated_heatmap(
    aggregated,
    title=f"{translations[lobby.value]} サブ使用率 [%]",
    figsize=(6, 7),
    use_annotation_image=True,
    annotation_image_zoom=0.55
)
plt.show()
```

```python
# スペシャル使用率

aggregated = a.aggregate_index_per_subject(players, "weapon-special", "usage-rate")
plt, ax = v.show_aggregated_heatmap(
    aggregated,
    title=f"{translations[lobby.value]} スペシャル使用率 [%]",
    figsize=(6, 7),
    use_annotation_image=True,
    annotation_image_zoom=0.55
)
plt.show()
```
